#### Задание №2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
from pylab import rcParams
rcParams["figure.figsize"] = 8, 6
```

```python
title_dict = {'fontsize': 18, 'fontweight':"bold", 'color' :'#808080', 'family' :'Calibri'}
label_dict = {'fontsize': 14, 'fontweight':"bold", 'color' :'#808080', 'family' :'Calibri'}
plt.plot(t, f, color="green")
plt.title("График f(t)", fontdict=title_dict)
plt.xlabel("Значения f", fontdict=label_dict)
plt.ylabel("Значения t", fontdict=label_dict)
plt.axis([-0.5, 9.5, -2.5, 2.5])
plt.show()
```

#### Задание №3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_ylabel('y1')
ax1.set_xlabel('x')
ax1.set_title('График y1', fontdict=title_dict)
ax2.plot(x, y2)
ax2.set_ylabel('y2')
ax2.set_xlabel('x')
ax2.set_title('График y2', fontdict=title_dict)
ax3.plot(x, y3)
ax3.set_ylabel('y3')
ax3.set_xlabel('x')
ax3.set_title('График y3', fontdict=title_dict)
ax4.plot(x, y4)
ax4.set_ylabel('y4')
ax4.set_xlabel('x')
ax4.set_title('График y4', fontdict=title_dict)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.4)
# я немного посвоевольничал и добавил название оси х, поэтому увеличил вертикальное расстояние между графиками
```
